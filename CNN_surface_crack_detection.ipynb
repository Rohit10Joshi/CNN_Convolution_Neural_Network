```python
# importing all the necessary libraries
import opendatasets as od
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D, MaxPooling2D, Flatten
import cv2
import numpy as np
```

```python
od.download("https://www.kaggle.com/datasets/arunrk7/surface-crack-detection")
```

```python
import os
negativeimages = os.listdir("/content/surface-crack-detection/Negative")  # importing negative images
print(len(negativeimages))
```

```python
negativeimages = negativeimages[:5000]  #taking only 5000 images out of 20000 for training
print(len(negativeimages))
```

Data/image preprocessing

```python
images = []
labels = []
for image in negativeimages:   # taking image from neagativeimages folders
  print(image)
  print("===")
```

```python
images = []
labels = []
negative_images_folder_path = "/content/surface-crack-detection/Negative/"
for image in negativeimages:
  img_path = os.path.join(negative_images_folder_path, image) # joining the path to the images in neagativeimages folder
  print(img_path)
  print("===")
```

```python
import cv2
images = []
labels = []
negative_images_folder_path = "/content/surface-crack-detection/Negative/"
for image in negativeimages:
  img_path = os.path.join(negative_images_folder_path, image)
  img = cv2.imread(img_path)
  img = cv2.resize(img, (64,64))   # resizing the images
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # converting images from BGR to RGB colour
  images.append(img)
  labels.append(0)   # 0 label for negative images
```

```python
# repeating the process for positive images
import os
positiveimages = os.listdir("/content/surface-crack-detection/Positive/")
positiveimages = positiveimages[:5000]
positive_images_folder_path = "/content/surface-crack-detection/Positive/"
for image in positiveimages:
  img_path = os.path.join(positive_images_folder_path, image)
  img = cv2.imread(img_path)
  img = cv2.resize(img, (64,64))
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  images.append(img)
  labels.append(1)
```

```python
len(images)
```

```python
len(labels)
```

```python
import numpy as np

images = np.array(images)
labels = np.array(labels)
```

Splitting data for training and testing

```python
xtrain,xtest,ytrain,ytest = train_test_split(images,labels,test_size = 0.2, random_state=53)
print(f'xtrain shape - ', xtrain.shape)
print(f'xtest shape - ', xtest.shape)
```

```python
xtrain
```

```python
xtrain = xtrain / 255.0
print(xtrain)
xtest = xtest / 255.0
```

Creating and training the CNN model

```python
model = Sequential() # brain without neurons
#Conv2D KERNELS LAYERS
#MaxPooling2D MAXPOOLING LAYERS
#Flatten MATRIX => 1D ARRAY(1D VECTOR)

model.add(Conv2D(32,(3,3), activation = 'relu', input_shape = (64,64,3))) #INPUT LAYER
model.add(MaxPooling2D(2,2))
# first cnn block

model.add(Conv2D(64,(3,3), activation = 'relu'))
model.add(MaxPooling2D(2,2))
# 2nd cnn block

model.add(Conv2D(64,(3,3), activation = 'relu'))
model.add(MaxPooling2D(2,2))
# 3rd cnn block

model.add(Conv2D(64,(3,3), activation = 'relu'))

model.add(Flatten())
#ann
model.add(Dense(64, activation = "relu")) #fcl
model.add(Dense(1,activation = "sigmoid")) #output layer

model.compile(optimizer = "adam", loss = "binary_crossentropy" , metrics = ["accuracy"])

model.fit(xtrain,ytrain, epochs = 3 , batch_size = 32 ,validation_data = (xtest,ytest))
```

Checking prediction on positive cracked image

```python
import cv2
import numpy as np

img_path = "/content/surface-crack-detection/Positive/00003.jpg"
image = cv2.imread(img_path)
image = cv2.resize(image,(64,64))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = np.expand_dims(image, axis = 0)
image = image / 255.0
prediction = model.predict(image)
if prediction[0][0] >= 0.5:
  print("1 => positive => crack detected")
else:
  print("0 => negative => crack not detected")
```

Evaluating models performance

```python
model.evaluate(xtest,ytest)
```
